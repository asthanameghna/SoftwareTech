```python
import sqlite3
import pandas
import numpy
from datetime import datetime

# Create the connection
db = sqlite3.connect(r'products.db')

# create the dataframe from a query
itemsdf = pandas.read_sql_query("SELECT * FROM items", db)
itemsdf.head()
```

```python
productsdf = pandas.read_sql_query("SELECT * FROM products", db)
productsdf.head()
```

```python
transactionsdf = pandas.read_sql_query("SELECT * FROM transactions", db)

# dayofyear
days = []
for d in range(len(transactionsdf['timestamp'])):
    date = itemsdf['timestamp'][d]
    datetime_object = datetime.strptime(date,"%Y-%m-%d %H:%M:%S")
    day = datetime_object.timetuple().tm_yday
    days.append(day)   
    
transactionsdf['dayofyear'] = days
transactionsdf.head()
```

```python
# dayofyear
days = []
for d in range(len(itemsdf['timestamp'])):
    date = itemsdf['timestamp'][d]
    datetime_object = datetime.strptime(date,"%Y-%m-%d %H:%M:%S")
    day = datetime_object.timetuple().tm_yday
    days.append(day)   
    
itemsdf['dayofyear'] = days


# price
prices = []
for p in itemsdf['productid']:
    price = productsdf['price'][p-1]
    prices.append(price/100.0)
    
itemsdf['price'] = prices    
```

```python
def daily_product_freq(day):
    
    """The function takes nth day of the year as argument 
    and returns the frequency of each product sold on that 
    particular day as a dataframe. """
    
    day_sale = itemsdf['productid'][itemsdf['dayofyear']==day].values
    products_sold = numpy.unique(day_sale, return_counts=True)

    
    productsolddf = pandas.DataFrame(products_sold[0],columns=['productid'])
    names = (productsdf['name'][products_sold[0]-1].values).tolist()
    productsolddf['name'] = names
    productsolddf['frequency'] = products_sold[1]
    return productsolddf
    pass



day = 345
productsolddf = daily_product_freq(day)
print('No. of Products sold on the ', day, 'th of the year')
productsolddf
```

```python
def daily_total_sales(day):
    
    """The function takes nth day of the year as argument 
    and returns the total sale on that particular day. """
    
    day_sale = itemsdf['productid'][itemsdf['dayofyear']==day].values
    products_sold = numpy.unique(day_sale, return_counts=True)
    
    productsolddf = pandas.DataFrame(products_sold[0],columns=['productid'])
    price = (productsdf['price'][products_sold[0]-1].values).tolist()
    productsolddf['price'] = price
    productsolddf['frequency'] = products_sold[1]
    productsolddf['total'] = productsolddf['price']*productsolddf['frequency']
    total = sum(productsolddf['total'])/100.0
    return total
    pass

day = 345
total = daily_total_sales(day)
print('Total Sales on ', day, 'th of the year: Â£', total)
```

```python
# 1: Cash | 2: Debit | 3: Credit | -1: Voided | nan: Not recorded

def daily_transaction_type_count(day):
    
    """The function takes nth day of the year as argument 
    and returns the frequency of each type of payment on 
    that particular day as a dataframe"""
    
    methods = transactionsdf['methodid'][transactionsdf['dayofyear']==day].values
    methods_type = (numpy.unique(methods, return_counts=True)[0]).tolist()
    methods_count = (numpy.unique(methods, return_counts=True)[1]).tolist()
    
    methods_type = methods_type[1:4]
    methods_count = methods_count[1:4] 

    methodsdf = pandas.DataFrame(methods_type, columns=['Method Type ID'])
    methodsdf['Method Type'] = ['Cash', 'Debit', 'Credit']
    methodsdf['Frequency'] = methods_count
    return methodsdf
    pass

day =345
methodsdf = daily_transaction_type_count(day)
print('Transaction Methods frequency on the ', day, 'th of the year')
methodsdf
```

```python
# 0: Full Price | 1: Half Price | 2: Free | nan: Not recorded

def daily_discounts(day):
    
    """The function takes nth day of the year as argument 
    and returns the frequency of each type of deal 
    and discount on that particular day as a dataframe. """
    
    day_sale = itemsdf['productid'][itemsdf['dayofyear']==day].values
    day_disc = itemsdf['discount'][itemsdf['dayofyear']==day].values
    day_deal = itemsdf['deal'][itemsdf['dayofyear']==day].values
    

    discountdf = pandas.DataFrame(day_sale,columns=['productid'])
    names = (productsdf['name'][day_sale-1].values).tolist()
    discountdf['name'] = names
    discountdf['discount'] = day_disc
    discountdf['deal'] = day_deal
    return discountdf
    pass

day = 346
daily_disc = daily_discounts(day)
daily_disc
```

```python
def total_sales(productid):
    
    """The function takes product ID as argument and 
    returns the frequency of product sold on every 
    day of the recording period as a dataframe. """
    
    items = (numpy.unique(itemsdf['productid'].values, return_counts=True))[0]
    freq = (numpy.unique(itemsdf['productid'].values, return_counts=True))[1]
    freq_prod = freq[numpy.where(items==productid)]
    price = productsdf['price'][productid-1]
    total = freq_prod[0]*price/100.0
    tot_sales = pandas.DataFrame([productid], columns=['productid'])
    tot_sales['frequency'] = freq_prod
    tot_sales['price'] = price
    tot_sales['total in Â£'] = total
    return tot_sales
    pass


productid = 5
tot_sales = total_sales(productid)
tot_sales
```
